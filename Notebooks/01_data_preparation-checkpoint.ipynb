```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
CLEAN_DIR = 'C:\\Users\\sansk\\Downloads\\ai_noise_reduction\\data\\clean_testset_wav'
NOISY_DIR = 'C:\\Users\\sansk\\Downloads\\ai_noise_reduction\\data\\noisy_dataset_wav'
```

```python
def load_audio(file_path, sr=16000):
    audio, _ = librosa.load(file_path, sr=sr)
    return audio
```

```python
def plot_waveform(audio, title='Waveform'):
    plt.figure(figsize=(10, 2))
    librosa.display.waveshow(audio, sr=16000)
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
sample_file = os.listdir(NOISY_DIR)[0]
audio = load_audio(os.path.join(NOISY_DIR, sample_file))
plot_waveform(audio, 'Sample Noisy Audio')
```
